# HOME ZIP CODE

```python
grouped_zc = df.groupby('HOME-ZIPCODE')
grouped_zc
```

What type of object is grouped_zc?

```python
# DataFrameGroupBy
```

DataFrameGroupBy objects contain a groups attribute. Display the value of the groups attribute belonging to the grouped_zc object.
